```python
import astropy.coordinates as coord
import astropy.units as u
import numpy as np
from astropy import units as u
from astropy.coordinates import SkyCoord
import matplotlib.pyplot as plt

myf=open("3csources.txt")
#first 2 lines are info
myf.readline()
myf.readline()
newf=open("3csources_file.txt","w")
newf.write("COLOR RED\nPA SKY\n")
for line in myf:
    name = line.strip().split()[0]
    ra=":".join(line.strip().split()[5:8])
    dec=":".join(line.strip().split()[8:11])
    mycoord = coord.SkyCoord([ra + " " + dec], frame=coord.FK4, unit=(u.hourangle,u.deg))
    mycoord = mycoord.transform_to(coord.FK5)
    newf.write("DOT %3.2f %3.2f \n"%(mycoord.ra.deg[0],mycoord.dec.deg[0]))
newf.close()


f = open("3csources_file.txt","r")
#f.read()
f.readline()
f.readline()

ra=[]
dec=[]

for line in f:
    RA = line.strip().split()[1]
    ra.append(RA)
    DEC = line.strip().split()[2]
    dec.append(DEC)
    

ra_sub=[]
dec_sub=[]    
g = open("sources_subtraction.txt","r")
g.readline()
g.readline()

for line in g:
    RA = line.strip().split()[1]
    ra_sub.append(RA)
    DEC = line.strip().split()[2]
    dec_sub.append(DEC)
    

#print(len(ra), len(ra_sub))
#print(len(dec), len(dec_sub))


coordinates = SkyCoord(ra, dec, frame=coord.FK5, unit=(u.deg,u.deg))
subtracted_coordinates = SkyCoord(ra_sub, dec_sub, frame=coord.FK5, unit=(u.deg,u.deg))



#need to match 3csources with the sources in the subtracted image because there are more sources in the 3c catalogue
#idx are the indecies that get the closest matches
#d2d on-sky distance between the matches
#d3d real-space distance between the matches
#in our case d3d can be ignored because we didn't give a line-of-sight distance, so its value is not particularly useful. But d2d provides a good diagnosis of whether we actually have real matches:


%matplotlib notebook
match = coord.match_coordinates_sky(subtracted_coordinates, coordinates, nthneighbor=1)
print(match)

plt.hist(match, histtype='step')
plt.xlabel('separation')
plt.legend()
plt.tight_layout()
plt.show()

#blue bin is the index
#orange is the angle
#green is the distance
```
